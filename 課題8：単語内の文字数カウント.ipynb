```python
EngW = []
Alp = {'a':0, 'b':0, 'c':0, 'd':0, 'e':0, 'f':0, 'g':0, 'h':0, 'i':0, 'j':0, 'k':0, 'l':0, 'm':0, 'n':0, 'o':0, 'p':0, 'q':0, 'r':0, 's':0, 't':0, 'u':0, 'v':0, 'w':0, 'x':0, 'y':0, 'z':0}

while True:
    sWk = input("英単語を入力してください：")
    if sWk=="":
        break
    else:
        for w in list(sWk):
            for Aa in Alp.keys():
                if w==Aa:
                    Alp[Aa] += 1
        EngW.append(sWk)
           

# print(f"{Alp}")
# print(f"入力した英単語：{EngW}")
EngW.sort()
print(f"入力した英単語：{EngW}")
for Ab in Alp.keys():
    if Alp[Ab] > 0 :
        print(f"{Ab}が{Alp[Ab]}個ありました")

```
